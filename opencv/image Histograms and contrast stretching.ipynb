```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir('../data')
```

### use calcHist

```python
img = cv.imread('salma.jpg', 0)
#                    chanel  mask
hist = cv.calcHist(img, [0], None, [255], [0, 255])
plt.plot(hist)
plt.show()
```

```python
def show(img):
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
img = cv.imread('1.png')
grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
show(grey)
```
